```python
# 用户侧需要 Embedding 的特征
user_embedding_features = ['user_id', 'activity_level_encoded', 'favorite_genre_encoded', 'gender_encoded', 'occupation_encoded', 'age_encoded']

# 电影侧需要 Embedding 的特征
movie_embedding_features = ['movie_id', 'title', 'genres', 'popularity', 'year', 'is_old_movie']
```

```python
import numpy as np

def get_embedding_config_for_features(df, feature_list, embedding_dim=64):
    """
    为指定的特征列表生成 Embedding 配置，支持 list 和 ndarray 类型列。
    获取非序列特征的Embedding配置
    Args:
        df (pd.DataFrame): 包含特征的 DataFrame。
        feature_list (list): 需要进行 Embedding 的特征列名列表。
        embedding_dim (int): Embedding 的默认维度。

    Returns:
        dict: 包含特征名和对应 Embedding 配置的字典。
    """
    embedding_config = {}
    for col in feature_list:
        if col in df.columns:
            first_item = df[col].dropna().iloc[0]
            # 判断是否为可迭代的序列（list 或 ndarray）
            if isinstance(first_item, (list, np.ndarray)):
                flat_values = set()
                max_seq_len=0
                for row in df[col]:
                    if isinstance(row, (list, np.ndarray)):
                        flat_values.update(row)
                        max_seq_len=len(row)
                num_unique = len(flat_values)
                embedding_config[col] = {'vocab_size': num_unique + 1, 'embedding_dim': embedding_dim, 'seq_len': max_seq_len+1}
            else:
                num_unique = max(df[col]) # 不可选df[col].unique() 否则会数组越界
                embedding_config[col] = {'vocab_size': num_unique + 1, 'embedding_dim': embedding_dim, 'seq_len': 1}  # 非序列特征默认为1
        else:
            print(f"警告: 特征 '{col}' 不存在于 DataFrame 中。")
    return embedding_config


# 获取用户侧的 Embedding 配置
user_embedding_config = get_embedding_config_for_features(user_features, user_embedding_features, embedding_dim=64)
print("用户 Embedding 配置:")
print(user_embedding_config)

# 获取电影侧的 Embedding 配置
movie_embedding_config = get_embedding_config_for_features(movie_features, movie_embedding_features, embedding_dim=64)
print("\n电影 Embedding 配置:")
print(movie_embedding_config)
```

```python
# 存储配置信息
import json
def write_embedding_config_to_json(filename, embedding_config):
    """
    将 Embedding 配置以 JSON 格式写入本地文件。

    Args:
        filename (str): 输出 JSON 文件名。
        embedding_config (dict): 要写入的 Embedding 配置字典。
    """
    with open(filename, 'w') as f:
        json.dump(embedding_config, f, indent=4)
    print(f"Embedding 配置已写入到: {filename}")

# 写入 Embedding 配置到 JSON 文件
write_embedding_config_to_json('./config/user_embedding_config.json', user_embedding_config)
write_embedding_config_to_json('./config/movie_embedding_config.json', movie_embedding_config)
```
